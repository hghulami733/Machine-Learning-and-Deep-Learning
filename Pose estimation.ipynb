```python
import cv2
import mediapipe as mp
import time
import math
```

```python
mp_draw = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()
```

```python
cap = cv2.VideoCapture(r"E:\python\segmentation\Computer Vision\Computer Vision\Data\Pose.mp4")
prev_time = 0

while True:
    success, img = cap.read()
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    results = pose.process(rgb_img)
    #print(results.pose_landmarks)
    if results.pose_landmarks:
        mp_draw.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        for id, landmark in enumerate(results.pose_landmarks.landmark):
            h, w, c = img.shape
            print(id, landmark)
            cx, cy = int(landmark.x*w), int(landmark.y*h)
            cv2.circle(img, (cx, cy), 5, (255, 255, 0), cv2.FILLED)

    curr_time = time.time()
    fps = 1 / (curr_time - prev_time)
    prev_time = curr_time

    cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN,
               3, (255, 0, 255), 3)

    cv2.imshow("image", img)
    cv2.waitKey(1)
    #cv2.destroyAllWindows()
```

```python
class PoseDetector():
    def __init__(self, mode=False, upper_body=False, smooth_landmark=True,
                 min_detection_conf=0.5, min_tracking_conf=0.5):
        self.mode = mode
        self.upper_body = upper_body
        self.smooth_landmark = smooth_landmark
        self.min_detection_conf = min_detection_conf
        self.min_tracking_conf = min_tracking_conf

        self.mp_draw = mp.solutions.drawing_utils
        self.mp_pose = mp.solutions.pose
        self.pose = self.mp_pose.Pose(self.mode, self.upper_body, self.smooth_landmark,
                                     self.min_detection_conf, self.min_tracking_conf)

    def find_pose(self, img, draw=True):
        self.rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        self.results = self.pose.process(self.rgb_img)
        
        #print(results.pose_landmarks)
        if self.results.pose_landmarks:
            if draw:
                self.mp_draw.draw_landmarks(img, self.results.pose_landmarks, 
                                            self.mp_pose.POSE_CONNECTIONS)
        return img

    def get_position(self, img, draw=True):
        self.landmark = []
        if self.results.pose_landmarks:
            for id, landmark in enumerate(self.results.pose_landmarks.landmark):
                h, w, c = img.shape
                #print(id, landmark)
                cx, cy = int(landmark.x * w), int(landmark.y * h)
                self.landmark.append([id, cx, cy])
                if draw:
                    cv2.circle(img, (cx, cy), 5, (255, 255, 0), cv2.FILLED)

        return self.landmark

    def find_angle(self, img, p1, p2, p3, draw=True):

        # Get the landmarks
        # _, x1, y1 = self.landmark[p1]   OR   
        x1, y1 = self.landmark[p1][1:]
        x2, y2 = self.landmark[p2][1:]
        x3, y3 = self.landmark[p3][1:]

        # Calculate the Angle

        angle = math.degrees(math.atan2(y3 - y2, x3 - x2) - matn.atan2(y1 - y2, x1 - x2))
        print(angle)

        if angle < 0:
            angle += 360
        
        # Draw
        if draw:

            cv2.line(img, (x1, y1), (x2, y2), (255, 255, 255), 3)
            cv2.line(img, (x3, y3), (x2, y2), (255, 255, 255), 3)
            
            cv2.circle(img, (x1, y1), 10, (0, 0, 255), cv2.FILLED)
            cv2.circle(img, (x1, y1), 15, (0, 0, 255), 2)
            
            cv2.circle(img, (x2, y2), 10, (0, 0, 255), cv2.FILLED)
            cv2.circle(img, (x2, y2), 15, (0, 0, 255), 2)
            
            cv2.circle(img, (x3, y3), 10, (0, 0, 255), cv2.FILLED)
            cv2.circle(img, (x3, y3), 15, (0, 0, 255), 2)

            cv2.putText(img, str(int(angle)), (x2 - 50, y2 + 50),
                       cv2.FONT_HERSHEY_PLAIN, 2, (0, 255, 0), 2)

        return angle
```

```python
def main():
    cap = cv2.VideoCapture(r"E:\python\segmentation\Computer Vision\Computer Vision\Data\Pose.mp4")
    prev_time = 0
    detector = PoseDetector()
    
    
    while True:
        success, img = cap.read()
        img = detector.find_pose(img)
        #landmark = detector.get_position(img)
        landmark = detector.get_position(img, draw=False)
        print(landmark[14])

        if len(landmark) != 0:
            cv2.circle(img, (landmark[14][1], landmark[14][2]), 5, (255, 0, 0), cv2.FILLED)
        
        curr_time = time.time()
        fps = 1 / (curr_time - prev_time)
        prev_time = curr_time
    
        cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN,
                   3, (255, 0, 255), 3)
    
        cv2.imshow("image", img)
        cv2.waitKey(1)
    #cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
