```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
heart_data = pd.read_csv(r"E:\python\mathmatic\datas\heart.csv")
heart_data.head()
```

```python
X = heart_data.drop(columns='target', axis=1)
y = heart_data['target']
```

```python
X = np.asarray(X)
y = np.asarray(y)
```

```python
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier(random_state=0), LinearRegression()]
```

```python
def compare_models_cross_validation():
    for model in models:
        cv_score = cross_val_score(model, X, y, cv=5)
        mean_accuracy = sum(cv_score) / len(cv_score)
        mean_accuracy = mean_accuracy * 100
        mean_accuracy = round(mean_accuracy, 2)
        print(f"The accuracy for model {model} is {cv_score}")
        print(f"Finall accuracy of model {model} is {mean_accuracy} %")
        print("__________________________________________________________")
        
```

```python
compare_models_cross_validation()
```

```python
models_list = [LogisticRegression(max_iter=10000), SVC(), KNeighborsClassifier(), RandomForestClassifier(random_state=0)]
```

```python
model_hyperparameters = {
    'Log_reg_hyperparameters':{
        'C':[1, 5, 10, 20]
    },
    
        'svc_hyperparameters':{
            'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
            'C':[1, 5, 10, 20]
    },
    'knn_hyperparameters':{
        'n_neighbors': [3, 5, 10]
    },
    
    'random_forest_classifier': {
        'n_estimators': [10, 20, 50, 100]
    }
}
```

```python
model_keys = list(model_hyperparameters.keys())
model_keys
```

```python
def ModelSelection(models1, hyperparameters_dictionary):
    result = []
    i = 0
    for model in models1:
        key = model_keys[i]
        params = hyperparameters_dictionary[key]
        i += 1
        print(model)
        print(params)
        print("_______________________________________________")
        
        classifier = GridSearchCV(model, params, cv=5)
        classifier.fit(X, y)
        result.append({
            'model_used': model, 
            'highest_score': classifier.best_score_,
            'best_hyperparameters': classifier.best_params_
        })
    result_dataframe = pd.DataFrame(result, columns=['model_used', 'highest_score', 'best_hyperparameters'])
    return result_dataframe
```

```python
ModelSelection(models_list, model_hyperparameters)
```
