```python
import cv2
import mediapipe
import time

class Face_mesh():
    def __init__(self, static_image_mode=False,
                max_num_faces=2, 
                refine_landmarks=False,
                min_detection_confidence=0.5,
                min_tracking_confidence=0.5):

        self.static_image_mode = static_image_mode
        self.max_num_faces = max_num_faces
        self.refine_landmarks = refine_landmarks
        self.min_detection_confidence = min_detection_confidence
        self.min_tracking_confidence = min_tracking_confidence
        
        self.mp_draw = mp.solutions.drawing_utils
        self.mp_face_mesh = mp.solutions.face_mesh
        self.face_mesh = self.mp_face_mesh.FaceMesh(self.static_image_mode,
                                                    self.max_num_faces,
                                                    self.refine_landmarks,
                                                    self.min_detection_confidence,
                                                   self.min_tracking_confidence)
        self.draw_spect = self.mp_draw.DrawingSpec(thickness=1, circle_radius=1)

    def find_face_mesh(self, img, draw=True):
        
        self.rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        self.results = self.face_mesh.process(self.rgb_img)

        faces = []
        if self.results.multi_face_landmarks:
            
            for face_landmark in self.results.multi_face_landmarks:
                if draw:
                    self.mp_draw.draw_landmarks(img, face_landmark, self.mp_face_mesh.FACEMESH_FACE_OVAL,
                                          self.draw_spect, self.draw_spect)

                    face = []
                    for id, lm in enumerate(face_landmark.landmark):
                        #print(lm)
                        h, w, c = img.shape
                        x, y, z = int(lm.x * w), int(lm.y * h), int(lm.z * c)
                        cv2.putText(img, str(id), (x, y), cv2.FONT_HERSHEY_PLAIN,
                                   0.7, (255, 0, 0), 1)
                        #print(id, x, y, z)
                        face.append([x, y, z])
                    faces.append(face)

        return img, faces


def main():

    cap = cv2.VideoCapture(r"D:\GOD OF WAR 1\Telegram\video_2024-02-27_10-56-12.mp4")

    prev_time = 0

    detector = Face_mesh()
    while True:
        seccuss, img = cap.read()
        img, faces = detector.find_face_mesh(img)
        if len(faces) != 0:
            print(len(faces[0]))
            #print(faces[0])
        
        current_time = time.time()
        fps = 1 / (current_time - prev_time)
        prev_time = current_time
    
        cv2.putText(img, f"fps: {int(fps)}", (20, 70), cv2.FONT_HERSHEY_PLAIN,
                   2, (255, 0, 0), 3)
    
        cv2.imshow("image", img)
        cv2.waitKey(1)

if __name__ == "__main__":
    main()
```
