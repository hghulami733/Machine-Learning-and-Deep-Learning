```python
video = r"E:\python\segmentation\Computer Vision\Computer Vision\Data\Object-Detection-101\Videos\ppe-1.mp4"
#video = r"D:\GOD OF WAR 1\Telegram (1)\Dance.mp4"

cap = cv2.VideoCapture(video)
#cap.set(3, 1366)
#cap.set(4, 720)

#model = YOLO("yolov8n.pt")
model = YOLO(r"E:\python\segmentation\Computer Vision\Computer Vision\Data\Object-Detection-101\Project 3 - PPE Detection\ppe.pt")

class_names = ['Hardhat', 'Mask', 'NO-Hardhat', 'NO-Mask', 'NO-Safety Vest', 'Person', 'Safety Cone',
              'Safety Vest', 'machinery', 'vehicle']

my_color = (0, 0, 255)

while True:
    success, img = cap.read() 
    
    results = model(img, stream=True)

    for r in results:
        boxes = r.boxes
        for box in boxes:
            x, y, x1, y1 = box.xyxy[0]
            x, y, x1, y1 = int(x), int(y), int(x1), int(y1)

            #cvzone.cornerRect(img, (x, y, x1, y1))
            #cv2.rectangle(img, (x, y), (x1, y1), my_color, 3)
            conf = math.ceil((box.conf[0] * 100)) / 100

            cls = int(box.cls[0])

            current_class = class_names[cls]

            if conf > 0.5:
                
                if current_class == "NO-Hardhat" or current_class == "NO-Safety Vest" or current_class == "NO-Mask":
                    my_color = (0, 0, 255)
    
                elif current_class == "Hardhat" or current_class == "Safety Vest" or current_class == "Mask":
                    my_color = (0, 255, 0)
                    
                else:
                    my_color = (255, 0, 0)
    
                cvzone.putTextRect(img, f"{class_names[cls]} {conf}",
                                   (max(0, x), max(35, y)), scale=1, thickness=1,
                                   colorB=my_color,
                                   colorT=(255, 255, 255),
                                   colorR=my_color,
                                   offset=5
                                  )
                cv2.rectangle(img, (x, y), (x1, y1), my_color, 3)

    cv2.imshow("image", img)
    cv2.waitKey(1)
```
