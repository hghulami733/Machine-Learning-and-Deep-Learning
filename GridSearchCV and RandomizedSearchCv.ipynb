```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
import sklearn.datasets
```

```python
data1 = sklearn.datasets.load_breast_cancer()
data1
```

```python
data = pd.DataFrame(data1.data, columns = data1.feature_names)
data
```

```python
data.describe()
```

```python
data['label'] = data1.target
```

```python
data
```

```python
data['label'].value_counts()
```

```python
X = data.drop(columns='label', axis=1)
y = data['label']
```

```python
X = np.asarray(X)
y = np.asarray(y)
```

```python
model = SVC()
```

```python
parameters = {
    'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
    'C':[1, 5, 10, 20]
}
```

```python
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
best = classifier.best_params_
best
```

```python
highest = classifier.best_score_
highest
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
grid
```

```python
randomized = RandomizedSearchCV(model, parameters, cv=5) #RandomizedSearchCV
```

```python
randomized.fit(X, y)
```

```python
randomized.best_params_ , randomized.best_score_
```

```python
result2 = pd.DataFrame(randomized.cv_results_)
```
