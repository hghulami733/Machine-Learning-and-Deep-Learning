```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('E:\python\python\82a9f1c1c473d6585e750ad2e3c05a41-d42d226d0dd64e7f5395a0eec1b9190a10edbc03\Medical_Cost.csv')
data
```

```python
dataframe = pd.get_dummies(data)
dataframe
```

```python
X = dataframe.drop('charges', axis=1)
y = dataframe['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
len(X), len(X_train), len(X_test)
```

```python
tf.random.set_seed(42)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
model.compile(loss=tf.keras.losses.mae,
             optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
             metrics=['mae'])
history = model.fit(X_train, y_train, epochs=200, verbose=1)
```

```python
model.evaluate(X_test, y_test)
```

```python
pd.DataFrame(history.history).plot()
plt.ylabel('loss')
plt.xlabel('epochs')
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
ct = make_column_transformer((
MinMaxScaler(), ['age', 'bmi', 'children']),
(OneHotEncoder(handle_unknown='ignore'), ['sex', 'smoker', 'region']))
X = data.drop('charges', axis=1)
y = data['charges']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
ct.fit(X_train)
X_train_normal = ct.transform(X_train)
X_test_normal = ct.transform(X_test)
X_train.loc[0]
```

```python
tf.random.set_seed(42)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
model.compile(loss=tf.keras.losses.mae,
             optimizer=tf.keras.optimizers.Adam(),
             metrics=['mae'])
model.fit(X_train_normal, y_train, epochs=100)
```

```python
model.evaluate(X_test_normal, y_test)
```
