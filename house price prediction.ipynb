```python
X = house_price.drop(columns='price', axis=1)
y = house_price['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
model = XGBRegressor()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_train)
score = metrics.r2_score(pred, y_train)
score2 = metrics.mean_absolute_error(pred, y_train)
score2, score
```

```python
test_pred = model.predict(X_test)
score3 = metrics.r2_score(test_pred, y_test)
score4 = metrics.mean_absolute_error(test_pred, y_test)
score3, score4
```

```python
plt.scatter(y_train, pred)
plt.xlabel('Actual price')
plt.ylabel('Prediction price')
plt.title('Actual price vs prediction')
plt.show()
```
