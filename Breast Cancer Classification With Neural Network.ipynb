```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset
```

```python
breast_cancer = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
breast_cancer
```

```python
breast_cancer['Target'] = breast_cancer_dataset.target
```

```python
breast_cancer
```

```python
breast_cancer.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = breast_cancer.drop(columns = 'Target', axis=1)
y = breast_cancer['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standard_X_train = scaler.fit_transform(X_train)
X_test_transform = scaler.fit(X_test)
```

```python
import tensorflow as tf
```

```python
tf.random.set_seed(3)
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(30,)),
    tf.keras.layers.Dense(20, activation=tf.keras.activations.relu),
    tf.keras.layers.Dense(2, activation=tf.keras.activations.sigmoid)
])
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
             optimizer = tf.keras.optimizers.Adam(),
             metrics=['accuracy'])
```

```python
history = model.fit(standard_X_train, y_train, epochs=10, validation_split = 0.1)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['Training data', 'Validation data'], loc = 'lower right')
plt.show();
```
